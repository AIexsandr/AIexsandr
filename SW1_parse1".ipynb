```python
import csv
import chardet

# Определение кодировки файла
def get_file_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

# Чтение файла с автоматическим определением кодировки
file_path = '/content/ML__2023-05-26T14-57_bewswa1.csv'
encoding = get_file_encoding(file_path)
print("Определенная кодировка файла:", encoding)

links = []

with open(file_path, 'r', encoding=encoding) as file:
    reader = csv.reader(file)
    next(reader)  # Пропускаем первую строку (заголовки)
    for row in reader:
        num1 = row[0]  # Значение 'num1' из текущей строки
        image_url = f"https://cdn.download.ams.birds.cornell.edu/api/v1/asset/{num1}/1200"


        # Дальнейшая обработка ссылки (например, сохранение изображения)
        image_url = image_url.split(',')[0] + '/1200'
        print(image_url)
        # Добавление ссылки в список
        links.append(image_url)

# Сохранение ссылок в файле
with open('/content/links.txt', 'w') as f:
    for link in links:
        f.write(link + '\n')
```

загрузить все фотографии из каталога /content/links.txt  вывести на печать и сохранить в папке /content/images изображению присвоить цифру взятую в строке линка перед /900 и расширение jpg

```python


import requests
import os

# Создание папки для сохранения изображений
image_dir = '/content/images-ml-1200'
os.makedirs(image_dir, exist_ok=True)

# Чтение ссылок из файла
with open('/content/links.txt', 'r') as file:
    links = file.readlines()

# Обработка ссылок
for link in links:
    link = link.strip()  # Удаление символов новой строки
    image_id = link.split('/')[-2]  # Получение ID изображения из ссылки
    image_path = os.path.join(image_dir, f"{image_id}.jpg")  # Формирование пути для сохранения изображения

    # Загрузка изображения
    response = requests.get(link)
    if response.status_code == 200:
        with open(image_path, 'wb') as f:
            f.write(response.content)
        print(f"Изображение '{image_id}.jpg' успешно сохранено.")
    else:
        print(f"Не удалось скачать изображение по ссылке: {link}")

print("Загрузка и сохранение изображений завершены.")
```

```python
import os
import requests

links_file = '/content/links.txt'
images_dir = '/content/images-ml-1200'

# Создание папки для сохранения изображений
os.makedirs(images_dir, exist_ok=True)

# Загрузка и сохранение изображений
with open(links_file, 'r') as f:
    for link in f:
        link = link.strip()
        image_id = link.split('/asset/')[1].split('/')[0]
        image_extension = 'jpg'
        image_path = os.path.join(images_dir, f"{image_id}.{image_extension}")

        response = requests.get(link)
        if response.status_code == 200:
            with open(image_path, 'wb') as image_file:
                image_file.write(response.content)
            print(f"Загружено и сохранено изображение: {image_path}")
        else:
            print(f"Не удалось загрузить изображение по ссылке: {link}")

```
